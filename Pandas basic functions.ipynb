Accessing values in dataframe 
There are 2 ways to access values in a dataframe.
1.Iloc() function
2.Loc() function

iloc[] function
It is used to access values in a dataframe. It is just like slicing a list to get values.

Allowed inputs are:
1. An integer, e.g. 5.
2. A list or array of integers, e.g. [4, 3, 0].
3. A slice object with ints, e.g. 1:7.
4. A boolean array.


```python
df.iloc[1:4] #it selects 3 columns and all the rows.
```

```python
df.iloc[1:4,2:4] #it selects the first 3 columns and 2 rows
```

loc() function
This is used to access the group of rows and column by label names and can be sliced using label names.

Allowed inputs are:
1. A single label, e.g. 5 or 'a', (note that 5 is interpreted as a label of the index, and never as an integer position along the index).
2. A list or array of labels, e.g. ['a', 'b', 'c'].
3. A slice object with labels, e.g. 'a':'f'.
4. A boolean array of the same length as the axis being sliced, e.g. [True, False, True].

```python
df.loc[1:4] # it selects the first 4 rows with the labels from 1 to 4.
```

```python
df.loc[1:4,'Name':'Ticket']# it selects the first 4 rows with the labels from 1 to 4. and selects all the columns from 'Name' to 'Ticket' 
```

inplace attribute is false by default so it will not change the dataframe. If made true there will be changes in the original dataframe.



```python
df.drop(0,inplace = True) #giving inplace true will be a permanent change in the dataframe.
df
```
