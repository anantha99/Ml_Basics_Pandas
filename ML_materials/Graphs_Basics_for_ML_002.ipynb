We use pyplot submodule from matplotlib for drawing graphs. <br> Pyplot makes matplotlib work like MATLAB.

## Important points to consider for a graph
It requires a pair of coordinates with x and y axis. <br>
we will declare them as arrays and they should have same size.<br>
If we want to display the graph we must make sure we use the pyplot.show()<br>

Importing pyplot from matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Matplotllib works with all sorts of data generally we will be using the NumPy arrays.

## plot() function 
If we want to show the graph as a line then we use the plt.plot().

```python
X = [2, 5, 7, 9]
Y = [13, 22, 37, 29]
plt.plot(X, Y) 
plt.show()
```

## Scatter() Function
plt.scatter() shows random points in the graph without connecting the points.

```python
X = [2, 5, 7, 9]
Y = [13, 22, 37, 29]
plt.scatter(X, Y) 
plt.show()
```

```python
#We can also add colour to the points
X = [2, 5, 7, 9]
Y = [13, 22, 37, 29]
plt.plot(X, Y,'ro') #here 'r' refers to red and 'o' refers to the shape of the point.
plt.show()
```

## Creating input usingÂ NumPy.
### numpy arange() function
arange() function helps in array creation based on numerical ranges.<br>
It creates an array with evenly spaced values and returns the reference of it.<br>

np.arange(start,stop,step,dtype=None)<br>

start - it specifies the first value in the array<br>
stop - it specifies the end but is not included in the array<br>
step - it specifies the spacing between 2 consecutive values in the array (default value is 1)<br>
dtype - type of elements in the output array(default is None)<br>

```python
t = np.arange(0.,5.,0.2)
plt.plot(t,t**3,color = 'green',marker = '^',linewidth = 2.5)
plt.show()
# color --> takes the color specified
# marker --> can be changed to any marker style like circle,diamond,triangle etc
# linewidth --> it defines the width of the line to be drawn
```

### Adding Multiple Lines To TheÂ Graph.
Multiple lines can be plotted in the same graph from the same plot() function.

```python
t = np.arange(0, 5, 0.2)
plt.plot(t,t**3,'ro',t,t**2,'bs',t,t,'ro')
plt.show()
```

### Creating Descriptive Graphs:
#### Adding Label andÂ Titles
To add labels to x and y-axis we use plt.xlabel() and plt.ylabel() respectively.<br>
To add titles we use plt.title() function

```python
X = [2, 5, 7, 9]
Y = [13, 22, 37, 29]
plt.plot(X, Y)
plt.xlabel("this is the xaxis")
plt.ylabel("this is the yaxis")
plt.title("This is the graph that you are looking at!")
plt.show()
```

## Adding scale to the graph
legend() function is used to set the scale for the graph but only after you specify the label() attribute in the plot() function definition.

```python
plt.plot([0.1, 0.2, 0.3, 0.4], [1, 2, 3, 4], label='first plot')
plt.plot([0.1, 0.2, 0.3, 0.4], [1, 4, 9, 16], label='second plot')
plt.legend()
plt.show()
```

### Adding grid to the graph 
Visualizing is made easy using the grid() function.

```python
plt.grid()
plt.show()
```

## Adding texts to the graph
While adding texts we should also specify the dimensions of the text.

```python
plt.text(0.5,0.9,"This is a text")
plt.show()
```
